#### Mengatasi Duplicate Data

```python
kolom = ['SO2', 'NO2', 'CO', 'O3', 'TEMP']
jumlah_duplikat = data.duplicated(subset=kolom).sum()

if jumlah_duplikat > 0:
    print(f"Ada {jumlah_duplikat} data yang duplikat dalam kolom yang dipilih. Namun karena ini data dari Alat Pengukur maka diabaikan")
else:
    print("Tidak ada data yang duplikat dalam kolom yang dipilih.")
data.to_csv('data_air_updated.csv', index=False)
```

#### Mengatasi Outlier

```python
kolom = ['SO2', 'NO2', 'CO', 'O3', 'TEMP']

for col in kolom:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

  
    data[col] = data[col].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))

data.to_csv('data_no_outlier.csv', index=False)
data.to_csv('data_air_updated.csv', index=False)
```

### Pertanyaan 4: Apakah pola polusi udara (SO2, NO2, dan CO) selama periode tertentu dapat memprediksi fluktuasi suhu di masa depan? (Hubungan antara kualitas udara dan ISPU)

```python
data = pd.read_csv('data_air_updated.csv')

# Gabungan kolom tahun, bulan, dan tanggal menjadi satu kolom 'Tanggal'
data['Tanggal'] = pd.to_datetime(data[['year', 'month', 'day']])

# Filter data selama 5 tahun (2013-2017)
start_date = '2013-01-01'
end_date = '2017-02-28'
data_filtered = data[(data['Tanggal'] >= start_date) & (data['Tanggal'] <= end_date)]


data.fillna(0, inplace=True)
# Mengubah kolom Pm10, SO2, NO2, CO, dan O3 menjadi tipe data numerik
data['PM10'] = pd.to_numeric(data['PM10'], errors='coerce')
data['SO2'] = pd.to_numeric(data['SO2'], errors='coerce')
data['NO2'] = pd.to_numeric(data['NO2'], errors='coerce')
data['CO'] = pd.to_numeric(data['CO'], errors='coerce')
data['O3'] = pd.to_numeric(data['O3'], errors='coerce')
# Menentukan parameter-parameter yang akan digunakan untuk menghitung ISPU
data['ISPU'] = (data['PM10'] + data['SO2'] + data['NO2'] + data['CO'] + data['O3']) / 5

# Selanjutnya, Anda dapat membuat kategori ISPU berdasarkan nilai-nilai ini jika diperlukan
data['Kategori_ISPU'] = pd.cut(data['ISPU'], bins=[0, 50, 100, 199, 299, data['ISPU'].max()],
                               labels=['Baik', 'Sedang', 'Tidak Sehat', 'Sangat Tidak Sehat', 'Berbahaya'])

# Pastikan bahwa kolom 'Kategori_ISPU' berisi data bertipe string
data['Kategori_ISPU'] = data['Kategori_ISPU'].astype(str)


plt.figure(figsize=(12, 6))
plt.plot(data['Tanggal'], data['PM10'], label='Pm10')
plt.plot(data['Tanggal'], data['SO2'], label='SO2')
plt.plot(data['Tanggal'], data['NO2'], label='NO2')
plt.plot(data['Tanggal'], data['CO'], label='CO')
plt.plot(data['Tanggal'], data['O3'], label='O3')
plt.xlabel('Tanggal')
plt.title('Hubungan Kualitas Udara dan Kategori ISPU (2013-2017)')
plt.legend()

# Tambahkan keterangan standar ISPU di bawah grafik
keterangan_standar = "ISPU Standar:\nBaik: 0-50\nSedang: 51-100\nTidak Sehat: 101-199\nSangat Tidak Sehat: 200-299\nBerbahaya: > 300"

# Letakkan teks keterangan_standar di bawah grafik
plt.text(0, -0.3, keterangan_standar, transform=plt.gca().transAxes, fontsize=10)
plt.show()

```

Selanjutnya kualitas udara akan di analisa baik atau buruknya berdasarkan nilai rata-rata dari masing masing parameter yang kemudian dilihat nilai ISPU nya. Pertama, kita mengonversi satuan dari ppb (parts per billion) ke μg/m³ (mikrogram per meter kubik), kita perlu mengetahui faktor konversi untuk masing-masing parameter. Faktor konversi ini bergantung pada massa jenis gas atau partikel yang bersangkutan. kita ambil contoh saja, berikut adalah faktor konversi umum untuk beberapa parameter:

1 ppb CO (Karbon Monoksida) = 1.145 μg/m³ |
1 ppb SO2 (Dioksida Belerang) = 2.62 μg/m³ |
1 ppb NO2 (Dioksida Nitrogen) = 1.88 μg/m³ |
1 ppb O3 (Ozon) = 2 μg/m³

```python
# Load data kualitas udara dan data parameter PM10, CO, SO2, NO2, O3
data_air_quality = pd.read_csv('data_air_updated.csv')

# Gabungkan kolom tahun, bulan, dan tanggal menjadi satu kolom 'Tanggal'
data_air_quality['Tanggal'] = pd.to_datetime(data_air_quality[['year', 'month', 'day']])

# Filter data berdasarkan rentang waktu (2013-2017)
start_date = '2013-01-01'
end_date = '2017-12-31'
data_filtered = data_air_quality[(data_air_quality['Tanggal'] >= start_date) & (data_air_quality['Tanggal'] <= end_date)]

# Group data berdasarkan tahun, bulan, dan parameter, lalu hitung rata-rata kualitas udara
grouped_air_quality = data_filtered.groupby([data_filtered['Tanggal'].dt.year, data_filtered['Tanggal'].dt.month])[['PM10', 'CO', 'SO2', 'NO2', 'O3']].mean()

# Ubah indeks agar lebih terbaca
grouped_air_quality.index.names = ['year', 'month']

# Tampilkan rata-rata parameter setiap bulan dari 2013-2017
print(grouped_air_quality)
grouped_air_quality.to_csv('rata_rata_parameter.csv', index=False)
```

Konversi Gas dari ppb ke μg/m³

```python
# Load data dari file CSV
data = pd.read_csv('rata_rata_parameter.csv')

# Menghitung μg/m³ untuk setiap parameter
def ppb_to_ug_per_m3(ppb, Mr, R, T):
    return ((ppb * Mr) / (R * T)) * (1 / 1000)

# Konstanta yang diperlukan untuk konversi
R = 8.314  # Konstanta gas (J/(mol·K))
Mr = {
    'PM10': 28.97,
    'CO': 28.01,
    'SO2': 64.06,
    'NO2': 46.01,
    'O3': 48.00
}

# Mengkonversi ppb ke μg/m³ untuk setiap parameter
for parameter in ['PM10', 'CO', 'SO2', 'NO2', 'O3']:
    if parameter != 'PM10':
        data[f'{parameter}_ug_per_m3'] = ppb_to_ug_per_m3(data[parameter], Mr[parameter], R, data['TEMP_kelvin'])

# Simpan data yang telah dikonversi ke dalam file CSV baru
data.to_csv('konversi_gas_parameter.csv', index=False)
```

```python
data = pd.read_csv('konversi_gas_parameter.csv')

# Membuat kolom ISPU untuk masing-masing parameter
parameters = ['PM10_ug_per_m3', 'CO_ug_per_m3', 'SO2_ug_per_m3', 'NO2_ug_per_m3', 'O3_ug_per_m3']

for parameter in parameters:
    data[f'{parameter}_ISPU'] = pd.cut(data[parameter], 
                     bins=[-1, 50, 100, 199, 299, float('inf')],
                     labels=['Baik', 'Sedang', 'Tidak Sehat', 'Sangat Tidak Sehat', 'Berbahaya'])

# Menyimpan data kualitas udara yang telah diberi kolom ISPU ke file CSV baru
data.to_csv('data_kualitas_udara_dengan_ISPU.csv', index=False)
```

Pertanyaan 1: Bagaimana tingkat polusi udara, yang diukur dengan PM10, SO2, NO2, dan CO, berhubungan dengan perubahan suhu seiring waktu?

Dalam analisis data, kita dapat menemukan bahwa hubungan antara tingkat polusi udara (diukur dengan PM10, SO2, NO2, dan CO) dan perubahan suhu seiring waktu cukup kompleks. Ada beberapa hubungan yang dapat diidentifikasi:

Terdapat beberapa korelasi positif antara tingkat polusi udara dan suhu, yang berarti ketika polusi meningkat, suhu juga cenderung meningkat.
Beberapa negatif korelasi juga terlihat, yang berarti ketika polusi meningkat, suhu cenderung menurun.
Namun, ada banyak faktor lain yang memengaruhi hubungan ini, seperti musim, cuaca, dan geografi.


Pertanyaan 2: Apakah terdapat korelasi antara kadar O3 dalam udara dengan fluktuasi suhu?

Analisis menunjukkan bahwa fluktuasi suhu tidak memiliki korelasi yang jelas dengan kadar O3 dalam udara. Oleh karena itu, tidak ada hubungan kuat antara kedua variabel ini dalam data yang dievaluasi.

Pertanyaan 3: Bagaimana pengaruh polutan udara (PM10, SO2, NO2, dan CO) pada suhu harian rata-rata?

Dari analisis data, kita dapat menyimpulkan bahwa polutan udara (PM10, SO2, NO2, dan CO) memiliki pengaruh yang kompleks pada suhu harian rata-rata. Hubungannya tergantung pada faktor-faktor lain seperti musim, cuaca, dan lokasi geografis. Terkadang tingkat polutan udara positif berkorelasi dengan peningkatan suhu, sementara terkadang mereka berkorelasi negatif dengan suhu. Pola ini dapat bervariasi seiring waktu.

Pertanyaan 4: Apakah pola polusi udara (PM10, SO2, NO2, dan CO) selama periode tertentu dapat memprediksi fluktuasi suhu di masa depan?

Dalam analisis ini, tidak ditemukan bukti yang kuat bahwa pola polusi udara (SO2, NO2, dan CO) selama periode tertentu secara konsisten dapat memprediksi fluktuasi suhu di masa depan. Fluktuasi suhu tergantung pada banyak faktor, dan polusi udara adalah salah satu faktor yang memengaruhi, tetapi tidak selalu merupakan prediktor yang andal. Prediksi suhu yang lebih akurat memerlukan pertimbangan lebih dari sekadar polusi udara, seperti data cuaca, geografi, dan faktor lingkungan lainnya. Nilai PM10 yang teranalisis adalah sedang hingga tidak sehat ini menandakan partikulat berakibat mulai penurunan pada jarak pandang dan menyebabkan jarak pandang turun secara signifikan, dan terjadi pengotoran debu di mana-mana.
